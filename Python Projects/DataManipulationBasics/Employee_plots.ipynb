# Employee Salary Visualization

This notebook visualizes employee salary data using histograms and boxplots. It provides insights into salary distribution and comparisons across departments.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load the Dataset


```python
try:
    df = pd.read_csv("Files/employee_data.csv")  # Attempt to load the employee data
except FileNotFoundError:
    print("Error: employee_data.csv not found")  # Handle the case where the file is missing
    # Create a dummy DataFrame for testing if the file does not exist
    df = pd.DataFrame({'Name': [], 'Department': [], 'Salary': [], 'Hire_Date': []})
```

## Create a Histogram of Salary Distribution


```python
# a) Create a histogram of the salary distribution
plt.figure(figsize=(8, 6))  # Set the figure size
plt.hist(df["Salary"], bins=10, edgecolor="black")  # Plot the histogram
plt.title("Salary Distribution")  # Set the title of the plot
plt.xlabel("Salary")  # Label for the x-axis
plt.ylabel("Frequency")  # Label for the y-axis
plt.grid(True, linestyle='--', alpha=0.6)  # Add grid lines for better readability
plt.show()  # Display the plot
```

## Create a Boxplot Showing Salary by Department


```python
# b) Create a boxplot showing Salary by Department
plt.figure(figsize=(10, 6))  # Set the figure size
sns.boxplot(x="Department", y="Salary", data=df)  # Create the boxplot
plt.title("Salary by Department")  # Set the title of the plot
plt.ylabel("Department")  # Label for the y-axis
plt.ylabel("Salary")  # Label for the y-axis
plt.grid(True, axis='y', linestyle='--', alpha=0.6)  # Add grid lines for better readability
plt.show()  # Display the plot
```
